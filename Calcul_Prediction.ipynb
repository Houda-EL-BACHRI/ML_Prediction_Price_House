```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Listings_num.csv')
```

```python
data
```

```python
y = data.price
y
```

```python
data.drop('price', inplace=True, axis=1)
```

```python
# Normaliser y

y_Train_scaled = (y-min(y))/(max(y)-min(y))
y_Train_scaled

```

les données numeriques possedent mauvaise correllation, heureusement il y'a les données categorielles qui possedent des informations predifctives


```python
data1 = pd.read_csv('/content/Listings_cat.csv')
data1
```

#Recodage des variables catégoriales en numériques

```python
data1 = pd.get_dummies(data1)
data1.head()
```

```python
data_final = pd.concat([data, data1], axis = 1)
data_final
```

```python
X= data_final
```

```python
# Normaliser X
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
X_Train_scaled = sc.fit_transform(X)
X_Train_scaled
```

```python
# Split the data to train and test:
from sklearn import model_selection
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_Train_scaled, y_Train_scaled, test_size=0.2)
```

#Model Prediction

```python
#Sélection du modèle comme régression linéaire et entraînement du modèle:

from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
#Test des performances du modèle à l'aide du carré R ajusté
from sklearn.metrics import r2_score
print((round(r2_score(y_train,regression.predict(x_train)),2))*100)
```
